```python
import tkinter as tk
import customtkinter as customTk
from customtkinter.windows.ctk_tk import CTk
from decimal import Decimal
import os
from PIL import Image
from PIL import ImageTk, Image
import mysql.connector


# Connect to the database
db = mysql.connector.connect(
    host="127.0.0.1",
    user="root",
    password="Harini@1994",
    database="restaurant_project"
)

# Create a cursor object to interact with the database
cursor = db.cursor()

# Define frame stack
frame_stack = []


# function to navigate back to previous screen
def back_button():
    # Remove the current frame from the stack
    frame_stack.pop()
    # Get the previous frame from the stack
    previous_frame = frame_stack[-1]
    # Forget the current frame and pack the previous frame
    current_frame = master.pack_slaves()[0]
    current_frame.pack_forget()
    previous_frame.pack(fill="both", expand=True)
            

'''Customer flow'''

def customer_Info():
    # Clear the current frame
    current_frame = master.pack_slaves()[0]
    current_frame.pack_forget()
    
    # create frames
    cframe_1 = customTk.CTkFrame(master)
    cframe_1.pack(pady=20, padx=60, fill="both", expand=True)
    
    # Add the current frame to the stack
    frame_stack.append(cframe_1)
    
    # create back button widget
    cback_button = customTk.CTkButton(cframe_1, text="Back", command=back_button)
    cback_button.pack(side="bottom", pady=20, padx=50)

    
    customTk.CTkLabel(cframe_1, text="Customer_ID").pack(side="top")
    cid_entry = customTk.CTkEntry(cframe_1)
    cid_entry.pack(side="top", padx=10)

    def validate_cid(cid_entry):
        cursor.execute(f"SELECT * FROM Customer WHERE Customer_id='{cid_entry}'")
        result = cursor.fetchone()
        return result
    
    csubmit_button = customTk.CTkButton(cframe_1, text="Submit", command=lambda: customer_validation())
    csubmit_button.pack(side="bottom", pady=20, padx=200)
    
    def customer_validation():
        cid = cid_entry.get()
        result = validate_cid(cid)
            
        if result:
            # if customer ID exists, switch to the customer frame
            switch_to_Restaurant_options_frame(cid)
        else:
            switch_to_add_customer_frame(cid)
            

    def switch_to_add_customer_frame(cid_entry):
        # Clear the current frame
        current_frame = master.pack_slaves()[0]
        current_frame.pack_forget()
        
        add_customer_frame = customTk.CTkFrame(master)
        add_customer_frame.pack(pady=20, padx=60, fill="both", expand=True)
        
        # Add the current frame to the stack
        frame_stack.append(add_customer_frame)
        
        # create back button widget
        cback_button = customTk.CTkButton(add_customer_frame, text="Back", command=back_button)
        cback_button.pack(side="bottom", pady=20, padx=50)

            
        name_label = customTk.CTkLabel(add_customer_frame, text="Customer Name:")
        name_label.pack(pady=10, padx=10)
        name_entry = customTk.CTkEntry(add_customer_frame)
        name_entry.pack(pady=10, padx=10)
        
        # create label and entry field for customer phone number
        phone_label = customTk.CTkLabel(add_customer_frame, text="Phone Number:")
        phone_label.pack(pady=10, padx=10)
        phone_entry = customTk.CTkEntry(add_customer_frame)
        phone_entry.pack(pady=10, padx=10)
        
        # create label and entry field for customer email
        email_label = customTk.CTkLabel(add_customer_frame, text="Email:")
        email_label.pack(pady=10, padx=10)
        email_entry = customTk.CTkEntry(add_customer_frame)
        email_entry.pack(pady=10, padx=10)
        
        # create label and entry field for customer address
        address_label = customTk.CTkLabel(add_customer_frame, text="Address:")
        address_label.pack(pady=10, padx=10)
        address_entry = customTk.CTkEntry(add_customer_frame)
        address_entry.pack(pady=10, padx=10)
        
        # create button to submit customer details
        submit_button = customTk.CTkButton(add_customer_frame, text="Submit", command=lambda: add_customer(cid_entry))
        submit_button.pack(pady=10, padx=10)
        
        def add_customer(cid_entry):
            cursor.execute(f"INSERT INTO Customer (Customer_id, Customer_Name, CPhone_Number, CEmail, CAddress) VALUES ('{cid_entry}', '{name_entry.get()}', '{phone_entry.get()}', '{email_entry.get()}', '{address_entry.get()}')")
            db.commit()
            conf_add_cust = customTk.CTkLabel(add_customer_frame, text=f"Customer {cid_entry} is added successfully, click on Restaurant Options button to continue", font=("Helvetica", 14))
            conf_add_cust.pack(pady=10, padx=10)
            
        rest_op_button = customTk.CTkButton(add_customer_frame, text="Restaurant Options", 
                                            command=lambda: switch_to_Restaurant_options_frame(cid_entry))
        rest_op_button.pack(pady=10, padx=10)
            
    def switch_to_Restaurant_options_frame(cid_entry):
        # clear the current frame and pack the options frame
        current_frame = master.pack_slaves()[0]
        current_frame.pack_forget()
        
        restaurant_options_frame = customTk.CTkFrame(master)
        restaurant_options_frame.pack(pady=20, padx=60, fill="both", expand=True)
        
        # Add the current frame to the stack
        frame_stack.append(restaurant_options_frame)
        
        # create back button widget
        cback_button = customTk.CTkButton(restaurant_options_frame, text="Back", command=back_button)
        cback_button.pack(side="bottom", pady=20, padx=50)
        
        # create a label with customer name
        customer_name_label = customTk.CTkLabel(restaurant_options_frame, text=f"Welcome to our UNT Restaurant", font=("Helvetica", 16))
        customer_name_label.pack(pady=20, padx=60)
        
        # create buttons for Reservation, Order, Feedback
        reservation_button = customTk.CTkButton(restaurant_options_frame, text="Reservation", 
                                                command=switch_to_reservation_frame)
        reservation_button.pack(pady=10, padx=200)
        
        order_button = customTk.CTkButton(restaurant_options_frame, text="Order", command=switch_to_order_frame)
        order_button.pack(pady=10, padx=200)
        
        feedback_button = customTk.CTkButton(restaurant_options_frame, text="Feedback", command=switch_to_feedback_frame)
        feedback_button.pack(pady=10, padx=200)
        
    '''Reservation Flow'''
    
    def switch_to_reservation_frame():
        # Clear the current frame
        current_frame = master.pack_slaves()[0]
        current_frame.pack_forget()
    
        # Create new frame
        reservation_frame = customTk.CTkFrame(master)
        reservation_frame.pack(pady=20, padx=60, fill="both", expand=True)
        
        # Add the current frame to the stack
        frame_stack.append(reservation_frame)
        
        # create back button widget
        cback_button = customTk.CTkButton(reservation_frame, text="Back", command=back_button)
        cback_button.pack(side="bottom", pady=20, padx=50)
    
        # Add title label
        customTk.CTkLabel(reservation_frame, text="Select Zone for Reserving Table", font=("Helvetica", 16)).pack(pady=20)
        
        Zone_1_btn = customTk.CTkButton(reservation_frame, text="Zone 1", command=lambda: select_table(1)).pack(pady=10)
        Zone_2_btn = customTk.CTkButton(reservation_frame, text="Zone 2", command=lambda: select_table(2)).pack(pady=10)
        Zone_3_btn = customTk.CTkButton(reservation_frame, text="Zone 3", command=lambda: select_table(3)).pack(pady=10)
    
        # Add buttons for zones
        def select_table(zone):
            
            # Clear the current frame
            #reservation_frame.destroy()
            current_frame = master.pack_slaves()[0]
            current_frame.pack_forget()
        
            # Create new frame
            table_frame = customTk.CTkFrame(master)
            table_frame.pack(pady=20, padx=60, fill="both", expand=True)
            
            # Add the current frame to the stack
            frame_stack.append(table_frame)
            
            # create back button widget
            cback_button = customTk.CTkButton(table_frame, text="Back", command=back_button)
            cback_button.pack(side="bottom", pady=20, padx=50)
            
            # Add title label
            customTk.CTkLabel(table_frame, text=f"Choose Table in Zone {zone}", font=("Helvetica", 16)).pack(pady=20)
                
            # Get available tables in the selected zone
            cursor.execute(f"SELECT Table_id FROM `Table` WHERE TStatus='Available' AND Zone_id={zone}")
            available_table_ids = [str(row[0]) for row in cursor.fetchall()]
            
            if available_table_ids:
                # Add a combobox to display the available table IDs
                table_id_combo_box = customTk.CTkComboBox(table_frame, values=available_table_ids)
                table_id_combo_box.pack(pady=20)
            else:
                no_tables_label = customTk.CTkLabel(table_frame, text="No tables available in this zone for Reservation! Click on Okay button to move back to Restaurant Options!", font=("Helvetica", 16))
                no_tables_label.pack(pady=20)
    
            def reserve_table():                             
                table_no = table_id_combo_box.get()
                
                # Get the next consecutive Reservation id from the Reservation table
                cursor.execute("SELECT MAX(Reservation_id) FROM Reservation")
                result = cursor.fetchone()
                reservation_id = int(result[0]) + 1 if result[0] else 1
    
                # Insert reservation into Reservation table
                cid_int = int(cid_entry.get())
                cursor.execute(f"INSERT INTO Reservation (Reservation_id, Customer_id, Restaurant_id, Table_id, R_Date, Time, Status) VALUES ('{reservation_id}', '{cid_int}', '1', '{table_no}', NOW(), NOW(), 'confirmed')")
                db.commit()
            
                # Update table status to 'occupied'
                cursor.execute(f"UPDATE `Table` SET TStatus='occupied' WHERE Table_id='{table_no}'")
                db.commit()
                
            
            # Add submit button
            submit_btn = customTk.CTkButton(table_frame, text="Submit", command=lambda: 
                                            ((reserve_table(),switch_to_Restaurant_options_frame(cid_entry)) 
                                             if available_table_ids else switch_to_Restaurant_options_frame(cid_entry))).pack()



    '''Order Flow'''
    
    def switch_to_order_frame():
        # Clear the current frame
        current_frame = master.pack_slaves()[0]
        current_frame.pack_forget()
    
        # Create new frame
        order_frame = customTk.CTkFrame(master)
        # Pack the reservation frame
        order_frame.pack(pady=20, padx=60, fill="both", expand=True)
        
        # Add the current frame to the stack
        frame_stack.append(order_frame)
        
        # create back button widget
        cback_button = customTk.CTkButton(order_frame, text="Back", command=back_button)
        cback_button.pack(side="bottom", pady=20, padx=50)
        
        global cart_items
        cart_items = []
    
        # Add title label
        customTk.CTkLabel(order_frame, text="Enter Order details for a Customer", font=("Helvetica", 16)).pack(pady=20)
        
        cursor.execute("SELECT MAX(order_id) FROM `Order`")
        result = cursor.fetchone()
        order_id = int(result[0]) + 1 if result[0] else 1
        order_id_label = customTk.CTkLabel(order_frame, text=f"Order id : {order_id}").pack()
        
        # Get the list of all tables
        table_id_label = customTk.CTkLabel(order_frame, text="Table id:")
        table_id_label.pack(pady=10, padx=10)
        cursor.execute("SELECT Table_id FROM `Table`")
        table_ids = [str(row[0]) for row in cursor.fetchall()]
        table_id_combo_box = customTk.CTkComboBox(order_frame, values=table_ids)
        table_id_combo_box.pack(pady=20)
        
        # Get the list of all Customers
        customer_id_label = customTk.CTkLabel(order_frame, text="Customer id:")
        customer_id_label.pack(pady=10, padx=10)
        cursor.execute("SELECT Customer_id FROM Customer")
        customer_ids = [str(row[0]) for row in cursor.fetchall()]
        customer_id_combo_box = customTk.CTkComboBox(order_frame, values=customer_ids)
        customer_id_combo_box.pack(pady=20)
        
        # Get the list of all Discounts
        Discount_id_label = customTk.CTkLabel(order_frame, text="Discount id:")
        Discount_id_label.pack(pady=10, padx=10)
        cursor.execute("SELECT * FROM Discount")
        discounts = [str(row[0]) for row in cursor.fetchall()]
        discounts_combo_box = customTk.CTkComboBox(order_frame, values=discounts)
        discounts_combo_box.pack(pady=20)
    
        
        def switch_to_food_category_frame(): 
            # Clear the current frame
            current_frame = master.pack_slaves()[0]
            current_frame.pack_forget()

            # Create new frame
            food_category_frame = customTk.CTkFrame(master)
            food_category_frame.pack(pady=20, padx=60, fill="both", expand=True)
            
            # Add the current frame to the stack
            frame_stack.append(food_category_frame)
        
            # create back button widget
            cback_button = customTk.CTkButton(food_category_frame, text="Back", command=back_button)
            cback_button.pack(side="bottom", pady=20, padx=50)

            # Add title label
            customTk.CTkLabel(food_category_frame, text="Choose the food category", font=("Helvetica", 16)).pack(pady=20)
    
            # Get the list of all food categories
            cursor.execute("SELECT FCName FROM food_category")
            food_categories = [row[0] for row in cursor.fetchall()]

            # Create dynamic buttons for each food category
            for category in food_categories:
                food_category_button = customTk.CTkButton(food_category_frame, text=category, width=200, height=60, 
                                                          command=lambda cat=category: switch_to_menu_item_frame(cat))
                food_category_button.pack(pady=10, padx=200)
        
            
        order_details_submit_button = customTk.CTkButton(order_frame, text="Submit", 
                                                         command=lambda: (switch_to_food_category_frame()))
        order_details_submit_button.pack(pady=10, padx=200)
                
        
                
        def switch_to_menu_item_frame(food_category_choosen):
            # Clear the current frame
            current_frame = master.pack_slaves()[0]
            current_frame.pack_forget()
    
            # Create new frame
            menu_item_frame = customTk.CTkFrame(master)
            menu_item_frame.pack(pady=20, padx=60, fill="both", expand=True)
            
            # Add the current frame to the stack
            frame_stack.append(menu_item_frame)
        
            # create back button widget
            cback_button = customTk.CTkButton(menu_item_frame, text="Back", command=back_button)
            cback_button.pack(side="bottom", pady=20, padx=50)
    
            # Add title label
            customTk.CTkLabel(menu_item_frame, text="Choose Menu Items", font=("Helvetica", 16)).pack(pady=20)
            
            # Fetch menu items from database
            cursor.execute(f"SELECT MName, Price FROM menu_item WHERE MName LIKE '%{food_category_choosen}%'")
            menu_items = [row for row in cursor.fetchall()]
            
            print(menu_items)
            
            # Create dynamic buttons for each food category
            for mitem in menu_items:
                menu_item_button = customTk.CTkButton(menu_item_frame, text=mitem[0], width=200, height=60, font=("Helvetica", 16), 
                                                      command=lambda mi=mitem: switch_to_ingredients_frame(mi))
                menu_item_button.pack(pady=10, padx=200)
                price_label = customTk.CTkLabel(menu_item_frame, text=f'{mitem[0]} price : {mitem[1]}', font=("Helvetica", 16))
                price_label.pack(pady=10, padx=200)
                
        
        def switch_to_ingredients_frame(menu_item_choosen):
            # clear previous frames
            current_frame = master.pack_slaves()[0]
            current_frame.pack_forget()

            # create new frame
            ingredients_frame = customTk.CTkFrame(master)
            ingredients_frame.pack(pady=20, padx=60, fill="both", expand=True)
            
            # Add the current frame to the stack
            frame_stack.append(ingredients_frame)
        
            # create back button widget
            cback_button = customTk.CTkButton(ingredients_frame, text="Back", command=back_button)
            cback_button.pack(side="bottom", pady=20, padx=50)
        
            ingredient_label = customTk.CTkLabel(ingredients_frame, text="Choose an ingredient:", font=("Helvetica", 16))
            ingredient_label.pack(pady=10)

            # get ingredients from database
            cursor.execute("SELECT IName, cost FROM Ingredient")
            ingredients = [row for row in cursor.fetchall()]
            
            x_pos = 10
            y_pos = 10
            
            none_name = "None"
            none_button = customTk.CTkButton(ingredients_frame, text=none_name, width=50, height=30, font=("Helvetica", 16), 
                                             command=lambda ing=none_name: switch_to_quantity_frame(menu_item_choosen, ing))
            none_button.place(x=x_pos, y=y_pos)
            y_pos += 35
            
            # Create dynamic buttons for each ingredient 
            for ing_item in ingredients:
                ing_item_button = customTk.CTkButton(ingredients_frame, text=ing_item[0], width=50, height=30, font=("Helvetica", 16), 
                                                     command=lambda ing=ing_item: switch_to_quantity_frame(menu_item_choosen, ing))
                ing_item_button.place(x=x_pos, y=y_pos)
                ing_price_label = customTk.CTkLabel(ingredients_frame, text=f'price : {ing_item[1]}', font=("Helvetica", 16))
                ing_price_label.place(x=x_pos+150, y=y_pos)
                y_pos += 35
                
            
        def switch_to_quantity_frame(menu_item_choosen, ingredient_choosen):
            # clear previous frames
            current_frame = master.pack_slaves()[0]
            current_frame.pack_forget()

            # create new frame
            quantity_frame = customTk.CTkFrame(master)
            quantity_frame.pack(pady=20, padx=60, fill="both", expand=True)
            
            # Add the current frame to the stack
            frame_stack.append(quantity_frame)

            # create back button widget
            cback_button = customTk.CTkButton(quantity_frame, text="Back", command=back_button)
            cback_button.pack(side="bottom", pady=20, padx=50)
        
            menu_item_quantity_label = customTk.CTkLabel(quantity_frame, text=f"Choose Quantity of {menu_item_choosen} you want to add to cart :", font=("Helvetica", 16))
            menu_item_quantity_label.pack(pady=10)
        
            quantity_numbers = [str(i) for i in range(1, 11)]

            # create dropdown list for quantity
            menu_item_quantity_combo_box = customTk.CTkComboBox(quantity_frame, values=quantity_numbers)
            menu_item_quantity_combo_box.pack(pady=20)

            # function to add item to cart
            def add_to_cart(menu_item, ingredient, menu_item_quantity):
                # Convert menu_item_quantity to integer
                menu_item_quantity = int(menu_item_quantity)
                
                
                print(ingredient)
                print(type(ingredient))
            
                # Get price of Menu Item from database
                menu_item_price_query = f"SELECT Price FROM menu_item WHERE MName = '{menu_item[0]}'"
                cursor.execute(menu_item_price_query)
                m_price = cursor.fetchone()[0]
                
                if ingredient=="None":
                    i_price = 0
                else:
                    # Get cost of ingredient from database
                    ingredient_cost_query = f"SELECT Cost FROM Ingredient WHERE IName = '{ingredient[0]}'"
                    cursor.execute(ingredient_cost_query)
                    i_price = cursor.fetchone()[0]

                # calculate item cost
                item_cost = (m_price + i_price)*menu_item_quantity

                # add item to cart
                cart_items.append((menu_item[0], ingredient[0], menu_item_quantity, item_cost))
                
                items_added_cart_label = customTk.CTkLabel(quantity_frame, text="Items added to the cart successfully!", font=("Helvetica", 16))
                items_added_cart_label.pack(pady=10)

            # function to switch to cart frame
            def switch_to_cart_frame():
                # clear previous frames
                current_frame = master.pack_slaves()[0]
                current_frame.pack_forget()

                # create new frame
                cart_frame = customTk.CTkFrame(master)
                cart_frame.pack(pady=20, padx=60, fill="both", expand=True)
                
                # Add the current frame to the stack
                frame_stack.append(cart_frame)

                # create back button widget
                cback_button = customTk.CTkButton(cart_frame, text="Back", command=back_button)
                cback_button.pack(side="bottom", pady=20, padx=50)

                # display cart items
                total_cost = 0
                for item in cart_items:
                    item_label = customTk.CTkLabel(cart_frame, text=f"{item[0]} - {item[1]} x {item[2]}", font=("Helvetica", 16))
                    item_label.pack(pady=10)
                    price_label = customTk.CTkLabel(cart_frame, text=f"${item[3]:.2f}", font=("Helvetica", 16))
                    price_label.pack(pady=10)
                    total_cost += item[3]

                # display total cost
                total_cost_label = customTk.CTkLabel(cart_frame, text=f"Total Cost: ${total_cost:.2f}", font=("Helvetica", 16))
                total_cost_label.pack(pady=10)

                # create add more items button
                add_more_items_button = customTk.CTkButton(cart_frame, text="Add More Items", 
                                                           command=lambda: switch_to_food_category_frame())
                add_more_items_button.pack(pady=10)

                # create payment button
                payment_button = customTk.CTkButton(cart_frame, text="Payment", 
                                                    command=lambda: switch_to_payment_frame(total_cost))
                payment_button.pack(pady=10)
            
            # function to make payment
            def switch_to_payment_frame(total_amount):
                # clear previous frames
                current_frame = master.pack_slaves()[0]
                current_frame.pack_forget()

                # create new frame
                payment_frame = customTk.CTkFrame(master)
                payment_frame.pack(pady=20, padx=60, fill="both", expand=True)
                
                # Add the current frame to the stack
                frame_stack.append(payment_frame)

                # create back button widget
                cback_button = customTk.CTkButton(payment_frame, text="Back", command=back_button)
                cback_button.pack(side="bottom", pady=20, padx=50)
                
                discount = discounts_combo_box.get()
                discount = int(discount)
                
                amount_label = customTk.CTkLabel(payment_frame, text=f"Total Amount: ${total_amount:.2f}", font=("Helvetica", 16))
                amount_label.pack(pady=10)
                
                discount_label = customTk.CTkLabel(payment_frame, text=f"Discount : ${discount}", font=("Helvetica", 16))
                discount_label.pack(pady=10)
                
                final_amount = total_amount - discount
                
                # Create a label to display the total amount
                final_amount_label = customTk.CTkLabel(payment_frame, text=f"Total Amount to be paid : {final_amount}", font=("Helvetica", 16))
                final_amount_label.pack(pady=10)
                
                # Create radio buttons for different payment modes
                payment_modes = ["Cash", "Credit Card", "Debit Card", "Others"]
                payment_mode_combo_box = customTk.CTkComboBox(payment_frame, values=payment_modes)
                payment_mode_combo_box.pack(pady=20)
                
                # Submit button
                payment_submit_button = customTk.CTkButton(payment_frame, text="Submit", 
                                                           command=lambda: (update_database_after_payment()))
                payment_submit_button.pack(pady=10, padx=200)
                
                def update_database_after_payment():
                    # Update Order table
                    cursor.execute(f"INSERT INTO `Order` (Order_id, Table_id, Customer_id, Discount_id, Time, Status) VALUES ('{order_id}', '{table_id_combo_box.get()}', '{customer_id_combo_box.get()}', '{discounts_combo_box.get()}', NOW(), 'Close')")
                    db.commit()
                    
                    # Update Payment table
                    cursor.execute("SELECT MAX(Payment_id) FROM Payment")
                    result = cursor.fetchone()
                    payment_id = int(result[0]) + 1 if result[0] else 1
                    cursor.execute(f"INSERT INTO Payment (Payment_id, Customer_id, Order_id, Order_Amount, Payment_Time, Payment_Method) VALUES ('{payment_id}', '{customer_id_combo_box.get()}', '{order_id}', '{final_amount}', NOW(), '{payment_mode_combo_box.get()}')")
                    db.commit()
                    
                    # Create a label to display the confirmation of Payment
                    payment_confirmation_label = customTk.CTkLabel(payment_frame, text="Amount Paid Successfully!", font=("Helvetica", 16))
                    payment_confirmation_label.pack(pady=10)
                    
                # Navigate back to Main Screen 
                return_back_button = customTk.CTkButton(payment_frame, text="Return Back", 
                                                        command=lambda: (switch_to_Restaurant_options_frame(customer_id_combo_box.get())))
                return_back_button.pack(pady=10, padx=200)
                
            
             # create add to cart button
            add_to_cart_button = customTk.CTkButton(quantity_frame, text="Add to Cart",
                                       command=lambda: add_to_cart(menu_item_choosen, ingredient_choosen,  menu_item_quantity_combo_box.get()))
            add_to_cart_button.pack(pady=20)

            # go to cart frame button
            go_to_cart_button = customTk.CTkButton(quantity_frame, text="Go to Cart", command=switch_to_cart_frame)
            go_to_cart_button.pack(pady=10)
            
            

    '''FeedBack Flow'''
    
    def switch_to_feedback_frame():
        # Clear previous frames
        current_frame = master.pack_slaves()[0]
        current_frame.pack_forget()

        # Create new frame
        feedback_frame = customTk.CTkFrame(master)
        feedback_frame.pack(pady=20, padx=60, fill="both", expand=True)
        
        # Add the current frame to the stack
        frame_stack.append(feedback_frame)
        
        # create back button widget
        cback_button = customTk.CTkButton(feedback_frame, text="Back", command=back_button)
        cback_button.pack(side="bottom", pady=20, padx=50)

        # Get max feedback_id from Feedback Table by query
        cursor.execute("SELECT MAX(Feedback_id) FROM Feedback")
        max_feedback_id = cursor.fetchone()[0]

        # Store +1 id to a feedback_id variable and display it as label
        feedback_id = max_feedback_id + 1
        feedback_id_label = customTk.CTkLabel(feedback_frame, text=f"Feedback ID: {feedback_id}", font=("Helvetica", 16))
        feedback_id_label.pack(pady=5)

        # Display combobox for Customer_id from Customer table query
        cursor.execute("SELECT Customer_id FROM Customer")
        customer_ids = [str(row[0]) for row in cursor.fetchall()]
        customer_label = customTk.CTkLabel(feedback_frame, text="Customer ID: ", font=("Helvetica", 16))
        customer_label.pack(pady=5)
        customer_combobox = customTk.CTkComboBox(feedback_frame, values=customer_ids)
        customer_combobox.pack()

        # Display Restaurant id label as 1 and store 1 in a variable
        restaurant_id = 1
        restaurant_id_label = customTk.CTkLabel(feedback_frame, text=f"Restaurant ID: {restaurant_id}", font=("Helvetica", 16))
        restaurant_id_label.pack(pady=5)

        # Display radio buttons for Ratings (1,2,3,4,5)
        rating_label = customTk.CTkLabel(feedback_frame, text="Rating: ", font=("Helvetica", 16))
        rating_label.pack(pady=5)
        rating_var = tk.StringVar()
        rating_var.set("1")

        for i in range(1, 6):
            rating_radio_button = tk.Radiobutton(feedback_frame, text=str(i), width=2, variable=rating_var, value=str(i))
            rating_radio_button.pack(pady=5)

        # Display Comment Entry field to provide the comments
        comment_label = customTk.CTkLabel(feedback_frame, text="Comment: ", font=("Helvetica", 16))
        comment_label.pack(pady=5)
        comment_entry = customTk.CTkEntry(feedback_frame, width=150, height=15)
        comment_entry.pack()

        # Submit button, which calls a function where it stores these in feedback table
        def submit_feedback():
            customer_id = customer_combobox.get()
            rating = rating_var.get()
            comment = comment_entry.get()

            # Store data in Feedback table
            cursor.execute(f"INSERT INTO Feedback (Feedback_id, Customer_id, Restaurant_id, Rating, Comment) VALUES ('{feedback_id}', '{customer_id}', '{restaurant_id}', '{rating}','{comment}' )")
            db.commit()

            # Display success message
            feedback_success_label = customTk.CTkLabel(feedback_frame, text="FeedBack stored successfully!", font=("Helvetica", 16))
            feedback_success_label.pack(pady=5)

        submit_button = customTk.CTkButton(feedback_frame, text="Submit", font=("Helvetica", 16), command=submit_feedback)
        submit_button.pack(pady=5)

        # Mainmenu button to return back to main_frame
        mainmenu_button = customTk.CTkButton(feedback_frame, text="Main Menu", font=("Helvetica", 16), 
                                             command=lambda: switch_to_Restaurant_options_frame(customer_combobox.get()))
        mainmenu_button.pack(pady=5, padx=10, side="bottom")


        
'''Employee Flow'''

def switch_to_Employee_frame():
    # Clear previous frames
    current_frame = master.pack_slaves()[0]
    current_frame.pack_forget()
    
    # create Employee side frame
    employee_side_frame = customTk.CTkFrame(master)
    employee_side_frame.pack(pady=20, padx=60,expand=True, fill="both")
    
    # Add the current frame to the stack
    frame_stack.append(employee_side_frame)

    # create back button widget
    cback_button = customTk.CTkButton(employee_side_frame, text="Back", command=back_button)
    cback_button.pack(side="bottom", pady=20, padx=50)
    
    def restaurant_back_data():
        # Clear previous frames
        current_frame = master.pack_slaves()[0]
        current_frame.pack_forget()

        # create Employee side frame
        rest_backdata_frame = customTk.CTkFrame(master)
        rest_backdata_frame.pack(pady=20, padx=60,expand=True, fill="both")

        # Add the current frame to the stack
        frame_stack.append(rest_backdata_frame)

        # create back button widget
        cback_button = customTk.CTkButton(rest_backdata_frame, text="Back", command=back_button)
        cback_button.pack(side="bottom", pady=20, padx=50)
        
        # create combobox for file selection
        rest_backdata_combobox = customTk.CTkComboBox(rest_backdata_frame, 
                                                      values=["Employee", "Kitchen_Station", "Feedback", "Supplier", "Purchase_Order", "Purchase_Item"])
        rest_backdata_combobox.pack(pady=10)
        
        def display_rest_data():
            rest_backdata_combo_name = rest_backdata_combobox.get()
            query = f"SELECT * FROM {rest_backdata_combo_name}"
            cursor.execute(query)
            results = cursor.fetchall()
            # clear any existing widgets in the frame
            for widget in rest_backdata_frame.winfo_children():
                widget.destroy()
            # create a label to display the column names
            cols = [column[0] for column in cursor.description]
            col_names = " | ".join(cols)
            col_label = customTk.CTkLabel(rest_backdata_frame, text=col_names)
            col_label.pack(pady=10)
            # create a label for each row of data
            for row in results:
                row_label = customTk.CTkLabel(rest_backdata_frame, text=" | ".join(str(cell) for cell in row))
                row_label.pack()
            cback_button = customTk.CTkButton(rest_backdata_frame, text="Back", command=back_button)
            cback_button.pack(side="bottom", pady=20, padx=50)
            
        rest_submit_button = customTk.CTkButton(rest_backdata_frame, text="Submit", command=display_rest_data)
        rest_submit_button.pack(pady=10)
        
    def customers_orders_data():
        # Clear previous frames
        current_frame = master.pack_slaves()[0]
        current_frame.pack_forget()

        # create Employee side frame
        customers_orders_frame = customTk.CTkFrame(master)
        customers_orders_frame.pack(pady=20, padx=60,expand=True, fill="both")

        # Add the current frame to the stack
        frame_stack.append(customers_orders_frame)

        # create back button widget
        cback_button = customTk.CTkButton(customers_orders_frame, text="Back", command=back_button)
        cback_button.pack(side="bottom", pady=20, padx=50)
        
        # create combobox for file selection
        customers_orders_data_combobox = customTk.CTkComboBox(customers_orders_frame, 
                                                              values=["Customer", "Table", "Reservation", "Discount", "Order", "Payment", "Feedback"])
        customers_orders_data_combobox.pack(pady=10)
        
        def display_customer_orders_data():
            customers_orders_data_combo_name = customers_orders_data_combobox.get()
            query = f"SELECT * FROM {customers_orders_data_combo_name}"
            cursor.execute(query)
            results = cursor.fetchall()
            # clear any existing widgets in the frame
            for widget in customers_orders_frame.winfo_children():
                widget.destroy()
            # create a label to display the column names
            cols = [column[0] for column in cursor.description]
            col_names = " | ".join(cols)
            col_label = customTk.CTkLabel(customers_orders_frame, text=col_names)
            col_label.pack(pady=10)
            # create a label for each row of data
            for row in results:
                row_label = customTk.CTkLabel(customers_orders_frame, text=" | ".join(str(cell) for cell in row))
                row_label.pack()
            cback_button = customTk.CTkButton(customers_orders_frame, text="Back", command=back_button)
            cback_button.pack(side="bottom", pady=20, padx=50)
            
        cust_submit_button = customTk.CTkButton(customers_orders_frame, text="Submit", command=display_customer_orders_data)
        cust_submit_button.pack(pady=10)

    # Create 2 buttons 
    restaurant_related_button = customTk.CTkButton(employee_side_frame, text="Restaurant Related", 
                                                   command=restaurant_back_data)
    restaurant_related_button.pack(pady=20)
    
    customer_orders_button = customTk.CTkButton(employee_side_frame, text="Customers Orders Data", 
                                                command=customers_orders_data)
    customer_orders_button.pack(pady=20)

   
        
                
        
master = customTk.CTk()
master.title("Restaurant Management Application")
master.geometry("1000x1000")
        
        
# create frames
main_frame = customTk.CTkFrame(master)
main_frame.pack(pady=20, padx=60, fill="both", expand=True)

# Add the main frame to the frame stack
frame_stack.append(main_frame)

# create labels and buttons
Employee_button = customTk.CTkButton(main_frame, text='Employee', command=switch_to_Employee_frame)
Employee_button.pack(pady=10, padx=200)
        
Customer_button = customTk.CTkButton(main_frame, text='Customer', command=customer_Info)
Customer_button.pack(pady=20, padx=200)

image_path = os.path.expanduser("C:\\Users\\klnv2\\OneDrive\\Desktop\\FDB\\Restaurant Project\\UI Code\\Test images")
image = customTk.CTkImage(Image.open(os.path.join(image_path, "Restaurant Image.jpg")), size=(1000, 1000))
label_image = customTk.CTkLabel(main_frame, image=image, text="UNT Restaurant", font=("Helvetica", 60), text_color="green")
label_image.pack(pady=5)


    
master.mainloop()



```
